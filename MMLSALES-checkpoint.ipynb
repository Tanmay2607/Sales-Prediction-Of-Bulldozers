```python
def get_number_of_unique_values(dataframe,column_names):
    df_dict={}
    for column in column_names:
        df_dict[column]=len(dataframe[column].unique())
    return df_dict
```

```python
def get_column_dtype(dataframe,columns):
    column_dtype={"Object" : [],"Numeric" : [],"Datetime" : []}
    for column in columns:
        if pd.api.types.is_numeric_dtype(dataframe[column]):
            column_dtype["Numeric"].append(column)
        elif pd.api.types.is_object_dtype(dataframe[column]):
            column_dtype["Object"].append(column)
        else:
            column_dtype["Datetime"].append(column)
    return column_dtype
```
